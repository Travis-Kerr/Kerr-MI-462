## How NULLs work when running within a Python notebook

You know from your last HW that the ```hearts_on_fire``` database has some NULL values within it. 

In fact, you discovered that the hearts_on_fire data has:

* 3 NULLs in the state column
* 683 NULLs in the class_year column
* 936 NULLs in the preferred_division column.

However, when looking at the data within a Jupyter notebook file, you won't see the word NULL, even though there are NULLS in the database.

Here's what will happen.  We use Python to request rows from the database.  When the records are returned, the notebook file will render/show the NULLs differently that they were shown in the database project.

So NULLs will instead appear in a Jupyter notebook as one of these values.

* NaN (Not a number)
* None

However, the SQL commands ```IS NULL``` and ```IS NOT NULL``` are still usable and still work just fine.  Run the next cell to prove this to yourself.


````python
#Here we put a SQL query between the triple quotes.
sql_statement_nulls = """ 
                          SELECT 1784 - COUNT(donation) AS missing_dontations, 
                                 1784 - COUNT(state) AS missing_states, 
                                 1784 - COUNT(affiliation) AS missing_affiliations, 
                                 1784 - COUNT(class_year) AS missing_class_years, 
                                 1784 - COUNT(preferred_division) AS missing_divisions, 
                                 1784 - COUNT(date_submitted) AS missing_submission_dates, 
                                 1784 - COUNT(donation_year) AS missing_donation_years
                            FROM hearts_on_fire;
                      """

#Run the query, passing it and the connection (conn) into the needed Python method.
#You get a dataframe (Python version of a table) back with the results of the query.
query_results = pd.read_sql_query(sql_statement_nulls, conn)

#Print the dataframe results.
query_results
````

In one table, show how many donors fall into the following categories for 2021.
